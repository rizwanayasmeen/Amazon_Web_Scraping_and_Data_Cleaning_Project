```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from selenium import webdriver
import seaborn as sns
from lxml import html
sns.set_style('darkgrid')
import math
%matplotlib inline
import matplotlib.pyplot as plt 

from pathlib import Path
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(ChromeDriverManager().install())


P_Name = []
P_Rating = []
P_Review = []
P_Original_price = []
P_Offer_price = []


for i in range(1,70):
    
    driver.get('https://www.amazon.in/s?k=laptop&page={}'.format(i))
    html=driver.page_source
    soup = BeautifulSoup(html,'html.parser')
    cards = soup.find_all('div', {"data-component-type":"s-search-result"})

    for card in cards:
        h2 = card.h2
        try:
            title = h2.text.strip()
        except:
            title = np.nan
        
        P_Name.append(title)


        try:
            price = card.find('span', class_='a-price-whole').text
        except:
            price = np.nan
        P_Offer_price.append(price)


        try:
            original_price = card.find('span', {'class': 'a-price a-text-price'}).find('span', {'class': 'a-offscreen'}).text.strip()
        except:
            original_price = np.nan
        
        P_Original_price.append(original_price)

        try:
            ratings = card.find('div', {'class':'a-row a-size-small'}).text
        except:
            ratings = np.nan
        P_Rating.append(ratings)

        try:
            no_of_reviews = card.find('a', {'class':'a-link-normal s-underline-text s-underline-link-text s-link-style'}).span.text
        except:
            no_of_reviews = np.nan
        P_Review.append(no_of_reviews)

        
        
        
        
df = pd.DataFrame({'Product' : P_Name, 'MRP': P_Original_price, 'Offer_Price' : P_Offer_price, 'Rating' : P_Rating, 'Reviews' : P_Review })        
            
driver.close()
  
      
```

```python
#temp varialbe r for striping and extracting reviews no's 
r = df1['Rating'].astype(str)
pro_reviews = r.apply(lambda z:z.split('(')[1:2])
df1['Reviews'] = pro_reviews
```

```python
df1
```

```python
df1['Reviews'] = df1['Reviews'].astype(str).str.replace("[", '')
df1['Reviews'] = df1['Reviews'].astype(str).str.replace("]", '')
df1['Reviews'] = df1['Reviews'].astype(str).str.replace(")", '')
df1['Reviews'] = df1['Reviews'].astype(str).str.replace(",", '')
df1['Reviews'] = df1['Reviews'].astype(str).str.replace("'", '')
```

```python
df1
```

```python
# Temp variable ra for extractring ratings
ra = r.str.replace('[a-zA-z]+', '')
pro_rating = ra.apply(lambda z:z.split(' ')[0])
pro_rating = pro_rating.str.slice(stop=3)
df1['Rating'] = pro_rating
```

```python
df1
```

```python
brand = df1['Product'].apply(lambda z:z.split()[0])
```

```python
#inserting brands at position 2 in dataframe
df1.insert(loc = 1,
          column = 'Brands',
          value = brand)
# show the dataframe
df1
```

```python
cols = ['MRP']
for col in cols:
    df1[col] = df1[col].map(lambda x: str(x).strip('₹')).astype(str)
    
df1['MRP'] = df1['MRP'].astype(str).str.replace(',', '')
df1['Offer_Price'] = df1['Offer_Price'].astype(str).str.replace(',', '')
df1 = df1.astype({'MRP':'float'})
df1 = df1.astype({'Offer_Price':'float'})
```

```python
pro_discount_value = df1['MRP'] - df1['Offer_Price']
pro_dic_percentage = (((df1['MRP'] - df1['Offer_Price'])/df1['MRP']) * 100)
pro_dic_percentage = round(pro_dic_percentage)
```

```python
df1["Discount Price"] = pro_discount_value
df1["Discount"] = pro_dic_percentage
```
